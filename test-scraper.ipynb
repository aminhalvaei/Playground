```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
```

```python
# Set up the Selenium WebDriver (using Chrome here)
driver = webdriver.Edge()

# Replace 'your_url' with the actual URL of the page you're scraping
driver.get("https://hesetazegi.ir/fa/recipes/fr-gheimeh-with-quince")
```

```python
try:
    # Wait for the "دستور پخت" tab to be clickable and click it
    cook_tab = WebDriverWait(driver, 10).until(
        EC.element_to_be_clickable((By.XPATH, "//div[contains(text(),'دستور پخت')]"))
    )
    driver.execute_script("arguments[0].click();", cook_tab)
    # Wait for all divs with class 'mt-4' to be present
    mt4_divs = WebDriverWait(driver, 10).until(
        EC.presence_of_all_elements_located((By.XPATH, "//div[@class='mt-4']"))
    )

    # Loop through each div with class 'mt-4'
    for mt4_div in mt4_divs:
        try:
            # Find the inner div that contains the h3 (title) and p (description)
            inner_div = mt4_div.find_element(By.XPATH, "./div[contains(@class, 'flex')]")

            # Extract the title from the h3 inside the inner div
            title = inner_div.find_element(By.XPATH, ".//h3[contains(@class, 'text-label-primary')]").text

            # Extract the description from the p tag
            description = inner_div.find_element(By.XPATH, ".//p[contains(@class, 'text-label-secondary')]").text

            # Print or store the title and description
            print(f"Title: {title}")
            print(f"Description: {description}")
            print("-" * 50)

        except Exception as e:
            print(f"Error processing a div: {e}")
            continue

finally:
    # Close the browser
    driver.quit()
```
